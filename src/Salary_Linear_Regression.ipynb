In this notebook we will be exploring the steps train a simple linear regression model to predict Salary based on Years of Experience

```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
#Loading the dataset
dataset = pd.read_csv('../dataset/Salary_Data.csv')
```

```python
#Check the datset top 5 rows
dataset.head()
```

```python
# Checking EDA
dataset.describe().T
```

```python
# Checking for missing values, data types and size of dataset
dataset.info()
```

```python
'''
We need to identify the dependent variable and independent variable. 
The Linear Regression model equation is y = mx + c 
As this is simple dataset, therefor we have 2 columns. YearsExperience as independent variable (x) and Salary as dependent variable (y) 
'''
X = dataset['YearsExperience']
y = dataset['Salary']
```

```python
''' 
fit() method of the LinearRegression model expects a 2D array for the feature inputs whreas we are passing 1D array in this case.
Reshape the input data to a 2D array using the reshape(-1, 1) method to satisfy the model's requirements.
'''

X = X.values.reshape(-1, 1)  
y = y.values.reshape(-1, 1)  
```

```python
''' 
In this setp we will be splitting the dataset into training and test dataset. From the output of previous cell we can see that there are 30 rows in the dataset.
Therefore we will take 20 records for the training set and 10 records for the test set.
The training dataset will be used for training model and test dataset will be used for testing the model.
''' 

from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 1/3, random_state=0)  
```

```python
'''
Building/Fitting a Simple Linear Regression model by feeding the data to the model.
Now we have the dataset ready. We will import the LinearRegression class from the scikit learn library.
We will be creating an object regressor of the class LinearRegression.
'''
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train)  
```

```python
# Predicting the Test set results 
y_pred= regressor.predict(X_test)
```

```python
# Visualizing the results for Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('YearsExperience vs Salary (Training set)')
```

```python
# Visualizing the results for Test set results
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('YearsExperience vs vs Salary (Test set)')
plt.show()
```
